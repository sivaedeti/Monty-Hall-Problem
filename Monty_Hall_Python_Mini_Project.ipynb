```python
#importing libraries
import random
import matplotlib.pyplot as plt
 
 
print("\t\tMonty Hall Problem Simulation Runner")
 
 #taking input for number of simulations
number_of_simulations = int(input("How many simulations do you want to run?: "))
 
printoption = input("Do you want to print simulations? y/n?")
 
 
 
#case 1: printing result without outputing individual simulation data
        
if printoption.lower()== 'n':
#sample size takes x-axis values for number of simulations ran
#win_percentage takes y-axis values for winning monty hall problem by switching doors
    sample_size = []
    win_percentage = []
    # running nested loop for getting each individual simulation data
    for j in range(1,number_of_simulations+1):
        #initializing trials_ran and times_won for each individual simulations
        trials_ran = 0
        times_won = 0
        for i in range(j):
            #print(f"Current Iteration: {i+1}")

            #setting monty_hall for actual values and dup_hall for duplicate
            monty_hall = [0,1,2]
            dup_hall = [0,1,2]

            #Door where car is behind
            car = random.randint(0,2)
            monty_hall.remove(car)

            #User taking his initial guess
            user_guess = random.randint(0,2)
            if user_guess in monty_hall:
                monty_hall.remove(user_guess)
 
            #host revealing his first non-car door
            reveal = monty_hall[0]
 
            dup_hall.remove(user_guess)
            dup_hall.remove(reveal)
 
            #user switching to remaining door
            switch = dup_hall[0]
            trials_ran += 1
 
            #Calculating how many times user won by only switching the door
            if switch == car:
                times_won += 1

        #Adding no.of trials done to sample_size
        #Adding winning percentage to win_percentage        
        sample_size.append(i)
        win_percentage.append(times_won/trials_ran)
    
    #plotting sample_size vs win_percentage
    plt.plot(sample_size, win_percentage)
    plt.title("Monty Hall Problem Simulation for switching of doors")
    plt.xlabel("Number of trials")
    plt.ylabel("Winning Percentage")
    plt.show()
        
        
else:
    sample_size = []
    win_percentage = []
    for j in range(1,number_of_simulations+1):
        trials_ran = 0
        times_won = 0
        for i in range(j):
            #print(f"Current Iteration: {i+1}")
            monty_hall = [0,1,2]
            dup_hall = [0,1,2]
 
            car = random.randint(0,2)
            monty_hall.remove(car)
 
            user_guess = random.randint(0,2)
            if user_guess in monty_hall:
                monty_hall.remove(user_guess)
 
            reveal = monty_hall[0]
 
            dup_hall.remove(user_guess)
            dup_hall.remove(reveal)
 
            switch = dup_hall[0]
 
 
            trials_ran += 1
 
            if switch == car:
                times_won += 1
                result = "Won"
            else:
                result = "Lost"
            print(f"Simulation number: {i+1}\nUser chose: {user_guess+1}\nHost revealed: {reveal+1}\nCar door: {car+1}\nUser switched to: {switch+1}\nResult: {result}\n\n")
        sample_size.append(i)
        win_percentage.append(times_won/trials_ran)
    plt.plot(sample_size, win_percentage)
    plt.show()
```
